```python
sum([x['count'] for x in rl])
```

```python
def list_to_count(rl):
    return sum([x['count'] for x in rl])
```

```python
list_to_count(ted.rating_list[0])
```

```python
ted=ted.assign(rating_counts=lambda x:x.rating_list.apply(list_to_count))
```

```python
ted.rating_counts
```
